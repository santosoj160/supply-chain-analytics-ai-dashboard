```python
missing_values = df.isnull().sum()

missing_values[missing_values > 0].sort_values(ascending=False)
```

```python
duplicates = df.duplicated().sum()

print(f"Duplicate Rows: {duplicates}")
```

```python
df = df.drop_duplicates()

print("Duplicates Removed")
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "")
    .str.replace(")", "")
)

df.columns
```

```python
date_columns = [col for col in df.columns if 'date' in col]

print(date_columns)
```

```python
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

print("Date Conversion Complete")
```

```python
df[date_columns].head()
```

```python
df = df.rename(columns={
    'order_date_dateorders': 'order_date',
    'shipping_date_dateorders': 'shipping_date'
})
```

```python
df[['order_date', 'shipping_date']].head()
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

Drop High Missing Columns

```python
df = df.drop(columns=[
    'product_description',
    'order_zipcode'
])

print("High Missing Columns Dropped")
```

Fill Remaining Missing Values

```python
#Numerical Columns - Fill with Median

num_cols = df.select_dtypes(include=np.number).columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
#Categorical Columns - Fill with 'Unknown'

cat_cols = df.select_dtypes(include='object').columns

df[cat_cols] = df[cat_cols].fillna('Unknown')
```

Check Missing Again

```python
#Check Missing Again
df.isnull().sum().sort_values(ascending=False).head(20)
```

Feature Engineering

Sekarang kita buat kolom baru yang sangat penting untuk analytics

```python
df['delivery_days'] = (
    df['shipping_date'] - df['order_date']
).dt.days
```

```python
df[['order_date', 'shipping_date', 'delivery_days']].head()
```

Tambahan Feature Engineering Professional
Profit Margin (%)

Ini bagus banget buat Tableau & insight bisnis.

```python
df['profit_margin'] = (
    df['benefit_per_order'] / df['sales_per_customer']
) * 100
```

Bersihkan Infinity / Error

Kadang ada pembagian 0.

```python
df['profit_margin'] = df['profit_margin'].replace(
    [np.inf, -np.inf],
    np.nan
)

df['profit_margin'] = df['profit_margin'].fillna(0)
```
